```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def lagrange(a, b, x, n, fun):
    P = 0
    h = (b-a)/n
    xi = a
    for i in range(n+1):
        L = 1
        xj = a
        for j in range(n+1):
            if i != j:
                L *= (x - xj)/(xi - xj)
            xj += h
        
        P += L * fun(xi)
        xi += h
    
    return P
```

```python
def fun(x):
    return 1/(1+x**2)
```

```python
lagrange(0, 2, np.array([0,math.pi/2]), 10, fun)
```

```python
n = 10
plt.figure()
x = np.linspace(-10, 10, 1000)
y_true = fun(x)
y_polynomial = lagrange(-10, 10, x, n-1, fun)
plt.plot(x, y_true)
plt.plot(x, y_polynomial)
```
