```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def divided_difference(x,xp,fp):
    n = len(xp)
    F = [[0]*n for _ in range(n)]
    for i in range(n):
        F[i][i] = fp[i]
    
    for l in range(1, n):
        for i in range(n-l):
            j = i + l
            F[i][j] = (F[i+1][j] - F[i][j-1])/(xp[j] - xp[i])

    P = F[0][0]
    tau = 1
    for j in range(1, n):
        tau *= x - xp[j-1]
        P += F[0][j] * tau
    
    return P
        
```

```python
def func(x):
    return np.sin(x)
```

```python
xp = np.linspace(0, 5, 10)
fp = func(xp)
```

```python
x = np.linspace(-2, 2, 1000)
y_polynomial = divided_difference(x, xp, fp)
y_true = func(x)

plt.figure()
plt.plot(x, y_true)
plt.plot(x, y_polynomial)
```
